> 자율주행 자동차는 4차 산업시대에 각광받는 분야 중 하나로 떠올랐고, 현재 Level 4인 고도 자율주행을 목표로 지속해서 개발하는 중입니다. 저는 ADAS 중에서도 자율주차 시스템이라는 기술에 주목했습니다. 일단 차량이 주행을 시작하면 언젠가는 정지해야 하므로, 완전 자율주행이라는 목표를 위해서 자율주차 시스템은 반드시 필요한 기능입니다.

> 저는 학부생 때, 자율주행 RC카에 기능을 추가하는 프로젝트를 수행하면서 자율주차 시스템에 한 번 벽을 느낀 경험이 있습니다. 운전자의 주행을 보조하는 차선이탈방지 시스템의 경우, 적외선 센서로 어렵지 않게 구현할 수 있었지만, 자율주차 시스템은 달랐습니다. 초음파 센서라는 눈을 이용해 전면주차와 후면주차를 구현하려고 했지만, 센서의 개수가 부족했습니다. 그래서 후면주차의 경우, 데모주행을 통해 구현할 수밖에 없었습니다. 또한, 전면주차마저도 주차공간의 벽을 발견한 이후에 주차공간 앞까지는 데모주행을 통해 들어가도록 구현하게 되었습니다. 이처럼 자율주차 시스템 구현의 어려움을 느끼게 되었고, 언젠가는 극복해보고 싶은 목표로 자리 잡게 되었습니다.

> 제가 머신러닝과 딥러닝 기술을 배우게 된다면 자율주차 시스템에 다시 한 번 도전하고 싶습니다. 시각적인 센서뿐만 아니라, 주변을 인식하는 센서를 사용해서 보이지 않는 곳의 거리 데이터도 수집할 수 있도록 학습하여 이를 구현하는 것입니다. 저는 머신러닝, 딥러닝 기술을 통해 Level 5인 완전 자율주행을 실현하는 미래도 머지않다고 생각합니다.
